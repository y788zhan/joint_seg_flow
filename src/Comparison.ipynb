## Train

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
split_10k_sgd = np.load("split_train_sgd_lambda_10k.npy")
split_1k_sgd = np.load("split_train_sgd_lambda1k.npy")
split_10k_adam = np.load("split_train_ADAM_lambda10k.npy")
split_1k_adam = np.load("split_train_ADAM_lambda1k.npy")
```

```python
nonsplit_10k_sgd = np.load("train_sgd_lambda10k.npy")
nonsplit_1k_sgd = np.load("train_sgd_lambda1k.npy")
nonsplit_10k_adam = np.load("train_ADAM_lambda10k.npy")
nonsplit_1k_adam = np.load("train_ADAM_lambda1k.npy")
```

```python
plt.plot(split_10k_sgd, label='split')
plt.plot(nonsplit_10k_sgd, label='non-split')
plt.title("SGD, 10,000")
plt.legend(loc='upper right')
# lr is 5e-6
```

```python
plt.plot(split_1k_sgd, label='split')
plt.plot(nonsplit_1k_sgd, label='non-split')
plt.title("SGD, 1,000")
plt.legend(loc='upper right')
#lr is 5e-5
```

```python
plt.plot(split_1k_adam, label='split')
plt.plot(nonsplit_1k_adam, label='non-split')
plt.title("ADAM, 1,000")
plt.legend(loc='upper right')
#lr is 1
```

```python
plt.plot(np.log(split_10k_adam), label='split')
plt.plot(np.log(nonsplit_10k_adam), label='non-split')
plt.title("ADAM, 10,000")
plt.legend(loc='upper right')
plt.ylabel("Logged Total Loss")
#lr is 1
```
